```python
import Tkinter
from PIL import Image, ImageTk,ImageFilter
import tkFileDialog
 
def changeImage():
    global im
    global img
    global filename
    myfiletypes = [('All files', '*'),('Python files', '*.py'),('Jpeg files','*.jpg')]
    #定義檔案類型
    filename = tkFileDialog.askopenfilename(filetypes=myfiletypes) 
    #透過tkFileDialog 開啟視窗供使用者選擇檔案
    im = Image.open( filename )
    #開啟Image
    img = ImageTk.PhotoImage(im)
    #ImageTk 透過此才能設定到Label上
    box1Label.configure(image=img)
    #設定label的 Image 
    #label就是新增一個圖文框，框框裡面只能放文字或圖片
    
def bigImage():
    global img
    global im 
    width = im.size[0]*4/3  #將寬度*4/3
    height = im.size[1]*4/3 
    #重新resize 大小
    im = im.resize( (width, height), Image.BILINEAR )
    img = ImageTk.PhotoImage(im)
    box1Label.configure(image=img)
    
def clear():
    global filename
    global img
    global im
    #重新開啟檔案
    im = Image.open( filename )
    img = ImageTk.PhotoImage(im)
    box1Label.configure(image=img)
    
def smallImage():
    global img
    global im 
    width = im.size[0]*1/3
    height = im.size[1]*1/3 #size減少為三分之一
    im = im.resize( (width, height), Image.BILINEAR )
    img = ImageTk.PhotoImage(im)
    box1Label.configure(image=img)
    
def rotateImage():
    global img
    global im 
    #旋轉圖片
    im = im.transpose( Image.ROTATE_90 )
    img = ImageTk.PhotoImage(im)
    box1Label.configure(image=img)

def blurImage():
    global img
    global im 
    #增加濾鏡
    im = im.filter( ImageFilter.BLUR )
    img = ImageTk.PhotoImage(im)
    box1Label.configure(image=img)
    
def contourImage():
    global img
    global im 
    im = im.filter( ImageFilter.CONTOUR )
    img = ImageTk.PhotoImage(im)
    box1Label.configure(image=img)
    
def embossImage():
    global img
    global im 
    im = im.filter( ImageFilter.EMBOSS )
    img = ImageTk.PhotoImage(im)
    box1Label.configure(image=img)
        
top = Tkinter.Tk()  #開新視窗


#用grid來定button的位置,用command把對應的fuction塞到button裡面
load = Tkinter.Button(top, text="上傳照片", command=changeImage)
load.grid(row=0, column=1)
load.config(width=25, height=1) #定義button大小

bigger = Tkinter.Button(top,text="照片放大", command=bigImage)
bigger.grid(row=1, column=1)
bigger.config(width=25, height=1)

smaller = Tkinter.Button(top,text="照片縮小", command=smallImage)
smaller.grid(row=2, column=1)
smaller.config(width=25, height=1)

rotate = Tkinter.Button(top,text="照片旋轉(90度)", command=rotateImage)
rotate.grid(row=3, column=1)
rotate.config(width=25, height=1)

blur= Tkinter.Button(top,text="濾鏡效果(模糊 BLUR)", command=blurImage)
blur.grid(row=4, column=1)
blur.config(width=25, height=1)

contour= Tkinter.Button(top,text="濾鏡效果(輪廓 CONTOUR)", command=contourImage)
contour.grid(row=5, column=1)
contour.config(width=25, height=1)

emboss= Tkinter.Button(top,text="濾鏡效果(浮雕 EMBOSS)", command=embossImage)
emboss.grid(row=6, column=1)
emboss.config(width=25, height=1)

clear = Tkinter.Button(top, text="清除所有效果", command=clear)
clear.grid(row=7, column=1)
clear.config(width=25, height=1)

box1Label = Tkinter.Label(top)
box1Label.grid(row=0, column=0 ,rowspan =8 )
 
Tkinter.mainloop()#跑迴圈讓程式不結束
```
